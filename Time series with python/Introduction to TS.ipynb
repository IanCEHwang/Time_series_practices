```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import pylab
```

### load data

```python
raw_csv = pd.read_csv("resources/Index2018.csv")
stock_data = raw_csv.copy()
```

### Examine data

```python
stock_data.head()
```

### Plot data

#### S&P 500

```python
raw_csv.spx.plot(figsize = (20 , 5),
                 title = "S&P500 Prices")
plt.show()
```

#### FTSE

```python
raw_csv.ftse.plot(figsize = (20 , 5),
                 title = "FTSE Prices")
plt.show()
```

```python
raw_csv.spx.plot(figsize = (20 , 5),
                 title = "S&P500 Prices")
raw_csv.ftse.plot(figsize = (20 , 5),
                 title = "FTSE Prices")
plt.title("S&P vs. FTSE")
plt.show()
```

### The QQ Plot

```python
scipy.stats.probplot(stock_data.spx , plot = pylab)
plt.show()
```

We see from the plot that data is not following the quantile line and jammed around 500, which means we can not use normal distribution related forecasting models.
