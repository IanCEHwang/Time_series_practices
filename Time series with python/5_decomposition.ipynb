```python
import pandas as pd
import numpy as np
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### prepare data
raw_csv = pd.read_csv("resources/Index2018.csv")
df_comp = raw_csv.copy()

### keep only one column
df_comp['date'] = pd.to_datetime(df_comp.date , dayfirst = True)
df_comp.set_index('date' , inplace = True)
df_comp['market_value'] = df_comp.spx
df_comp = df_comp[['market_value']]
df_comp = df_comp.asfreq('b')

df_comp['market_value'] = df_comp['market_value'].fillna(method = 'ffill')
```

### Seasonality

Additive seasonal decomposition

```python
s_dec_additive = seasonal_decompose(df_comp.market_value , model = 'additive')
s_dec_additive.plot()
plt.show()
```

Imply no seasonality in data

Multiplicative decomposition

```python
s_dec_additive = seasonal_decompose(df_comp.market_value , model = 'multiplicative')
s_dec_additive.plot()
plt.show()
```
